```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('/student_performance.csv')
df.describe()
```

```python
print("missing values per column: ",df.isnull().sum())
```

```python
x=df[['weekly_self_study_hours','attendance_percentage','class_participation']]
y=df['total_score']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print("Training data shape:", x_train.shape)
print("Testing data shape:", x_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_train)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
mse=mean_squared_error(y_train,y_pred)
rmse=np.sqrt(mse)
r2=r2_score(y_train,y_pred)
print(mse)
print(rmse)
print(r2)
```

In this project, a machine learning regression model was developed to predict student academic performance using factors such as weekly self-study hours, attendance percentage, and class participation.

The model achieved a satisfactory R² score, indicating that these factors play a significant role in determining student performance. The results highlight the importance of consistent study habits and regular class engagement in academic success.
